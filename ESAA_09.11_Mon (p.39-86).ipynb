### apply lambda 식으로 데이터 가공
- lambda 함수 - lambda 입력인자: 계산식   꼴로 정의한다.
- lambda 식을 이용할 때 보통 map()함수를 결합해서 사용한다.

```python
def get_square(a):
    return a**2
print('3의 제곱은',get_square(3))

lambda_sq = lambda x: x**2
print('3의 제곱은',lambda_sq(3))
```

```python
a=[1,2,3]
sq = map(lambda x:x**2,a)
list(sq) # 이 때 반드시 list()를 취해줘야 한다!
```

```python
t_df['Name_len']=t_df['Name'].apply(lambda x:len(x)) # Name칼럼의 값들의 길이를 Name_len 칼럼에 저장
t_df[['Name','Name_len']].head(3)
```

```python
t_df["Child_Adult"] = t_df["Age"].apply(lambda x : "Child" if x<=15 else "Adult")
t_df[["Age","Child_Adult"]].head(8)
```

```python
t_df["Child_Adult"] = t_df["Age"].apply(lambda x : "Child" if x<=15 else ("Adult" if x<=60 else "Elderly"))
t_df["Child_Adult"].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 작성
def clsfy(age):
    if age<=5:
        cat="baby"
    elif age<=15:
        cat="Child"
    elif age<=25:
        cat="Student"
    elif age<=35:
        cat = "Young Adult"
    elif age<=60:
        cat = "Adult"
    else:
        cat="Elderly"
    return cat

# lambda 식에 위에서 생성한 get_catefgory() 함수를 반환값으로 지정
# clsfy(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat값 반환
t_df['Age_cat'] = t_df["Age"].apply(lambda x: clsfy(x))
t_df[["Age","Age_cat"]].head()
```
