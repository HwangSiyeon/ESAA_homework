### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips= pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv")
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop("Unnamed: 0",axis=1)
```

### Step 5. Plot the total_bill column histogram

```python
ax=plt.subplot()
ax= sns.distplot(tips["total_bill"], kde=False)
ax.set_title('Histogram of Total Bill')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(tips['total_bill'], tips['tip'])
ax.set_title('Scatter Plot of Total Bill vs Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
fig = plt.figure()
ax =sns.relplot(x="total_bill", y="tip",data=tips, s=tips["size"]*10)
ax.fig.suptitle('Scatter Plot of Total Bill and Tip and Size')
ax.set_axis_labels(x_label='Total Bill',ylabel='Tip')
```
