```python
def calculate_volatility(data, window=20):
    """
    Calculate the rolling volatility using the standard deviation of returns.
    :param data: DataFrame with OHLCV data
    :param window: The number of periods to use for calculating the standard deviation
    :return: DataFrame with the volatility values
    """

    # Calculate daily returns
    data['returns'] = data['Close'].pct_change()

    # Calculate the rolling standard deviation of returns
    data['volatility'] = data['returns'].rolling(window=window).std()

    return data
```

```python
#newdata = pd.read_csv('/content/drive/MyDrive/BDA/newData.csv') # 시현
newdata = pd.read_csv('/content/drive/MyDrive/ESAA/BDA/newData.csv') # 수미
#newdata = pd.read_csv('/content/drive/MyDrive/newData.csv의 사본') #승원
data = pd.concat([data, newdata], axis=0)
data=calculate_volatility(data)
```

# 2. ta 라이브러리를 이용하여 보조지표 추가


ㅇ보조지표 리스트(추가 가능, 블로그 베껴온 것)
- SMA5, 20, 120 : Simple Moving Average
- EMA12, 26 : Exponential Moving Average
- MACD : Moving Average Convergence/Divergence
- RSI : Relative Strength Index
- BBANDS : Bollinger Bands

```python
import csv
import numpy as np
import talib
import pandas as pd

Close = np.array(df[['Close']])

df['sma5'] = talib.SMA(np.asarray(df['Close']), 5)
df['sma20'] = talib.SMA(np.asarray(df['Close']), 28)
df['sma126'] = talib.SMA(np.asarray(df['Close']), 120)

df['ema12'] = talib.EMA(np.asarray(df['Close']), 12)
df['ema26'] = talib.EMA(np.asarray(df['Close']), 26)

upper, middle, lower = talib.BBANDS(np.asarray(df['Close']), timeperiod=28, nbdevup=2, nbdevdn=2, matype=8)
df['dn'] = lower
df['mavg'] = middle
df['up'] = upper
df['pctB'] = (df.Close - df.dn)/(df.up - df.dn)

rsi14 = talib.RSI(np.asarray(df['Close']), 14)
df['rsi14'] = rsi14

macd, macdsignal, macdhist = talib.MACD(np.asarray(df['Close']), 12, 26, 9)
df['macd'] = macd
df['signal'] = macdsignal

# df.to_csv('/content/drive/MyDrive/ESAA/BDA/test1.csv',sep=',',na_rep='NaN')
```

```python
#일목균형표 합치기

# Ichimoku Cloud 지표 계산
tenkan_sen = talib.EMA(np.asarray(df['Close']), timeperiod=9)
kijun_sen = talib.EMA(np.asarray(df['Close']), timeperiod=26)
senkou_span_a = (tenkan_sen + kijun_sen) / 2
senkou_span_b = talib.EMA(np.asarray(df['Close']), timeperiod=52)
chikou_span = pd.Series(df['Close'].shift(-26), name='chikou_span')  # 26 기간 미래로 이동하여 Series로 변환

# DataFrame에 Ichimoku Cloud 지표 추가
df['tenkan_sen'] = tenkan_sen
df['kijun_sen'] = kijun_sen
df['senkou_span_a'] = pd.Series(senkou_span_a).shift(26)  # 26 기간 미래로 이동하여 Series로 변환
df['senkou_span_b'] = pd.Series(senkou_span_b).shift(26)  # 26 기간 미래로 이동하여 Series로 변환
df['chikou_span'] = chikou_span
```

```python
#Voltality 관련 변수 추가

#ATR (Average True Range) 계산
df['ATR'] = talib.ATR(np.asarray(df['High']), np.asarray(df['Low']), np.asarray(df['Close']), timeperiod=14)

# Keltner Channels (KC) 계산
df['KC_MIDDLE'] = talib.SMA(np.asarray(df['Close']), 20)
df['KC_UPPER'] = df['KC_MIDDLE'] + 2 * df['ATR']
df['KC_LOWER'] = df['KC_MIDDLE'] - 2 * df['ATR']

# Donchian Channels (DC) 계산
df['DC_UPPER'] = talib.MAX(np.asarray(df['High']), timeperiod=20)
df['DC_LOWER'] = talib.MIN(np.asarray(df['Low']), timeperiod=20)

# Ulcer Index (UI) 계산
df['UI'] = talib.ULTOSC(np.asarray(df['High']), np.asarray(df['Low']), np.asarray(df['Close']), timeperiod1=7, timeperiod2=14, timeperiod3=28)
```

```python
# momentum 관련 변수 추가
# RSI 계산
df['RSI'] = talib.RSI(df['Close'])

# SRSI 계산
slowk, slowd = talib.STOCHRSI(df['RSI'])
df['SRSI_K'] = slowk
df['SRSI_D'] = slowd


# Stochastic Oscillator(SR) 계산
slowk, slowd = talib.STOCH(df['High'], df['Low'], df['Close'])
df['Stoch_Oscillator_K'] = slowk
df['Stoch_Oscillator_D'] = slowd
df['SMA'] = talib.SMA(df['Close'])

# TSI 계산
def calculate_tsi(df):
  mom = talib.MOM(df['Close'], timeperiod=25)
  ema_mom = talib.EMA(mom, timeperiod=13)
  mom_of_mom = talib.MOM(ema_mom, timeperiod=1)
  tsi = 100 * (ema_mom / mom_of_mom)
  return tsi

df['TSI'] = calculate_tsi(df)

# UO 계산
df['UO'] = talib.ULTOSC(df['High'], df['Low'], df['Close'])

# Williams %R(WR) 계산
df['WR'] = talib.WILLR(df['High'], df['Low'], df['Close'])

# Awesome Oscillator(AO) 계산
def calculate_awesome_oscillator(df):
  median_price = (df['High'] + df['Low']) / 2
  short_moving_average = median_price.rolling(window=5, min_periods=1).mean()
  long_moving_average = median_price.rolling(window=34, min_periods=1).mean()
  ao = short_moving_average - long_moving_average
  return ao

df['AO'] = calculate_awesome_oscillator(df)

# KAMA 계산
df['KAMA'] = talib.KAMA(df['Close'])

# ROC 계산
df['ROC'] = talib.ROC(df['Close'])

# PPO 계산
df['PPO'] = talib.PPO(df['Close'])

# PVO 계산
df['PVO'] = talib.PPO(df['Volume'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 특성 중요도를 가져오고 중요도 크기순으로 정렬
feature_importances = rf_model2.feature_importances_
indices = np.argsort(feature_importances)[::-1]

# 중요도가 큰 순서로 특성 이름과 중요도를 출력
print("Feature ranking:")
for f in range(X_train.shape[1]):
    print(f"{X_train.columns[indices[f]]}: {feature_importances[indices[f]]}")

# 중요도를 시각화하여 플롯을 생성
plt.figure(figsize=(10, 6))
plt.bar(range(X_train.shape[1]), feature_importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), X_train.columns[indices], rotation=45)
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.title("Feature Importance Plot")
plt.show()
```
